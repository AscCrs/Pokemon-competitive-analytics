```python
import pandas as pd
```

```python
# Leer el CSV
df = pd.read_csv('../data_quiz_obtained/Pokémon  (Respuestas) - Respuestas de formulario 1.csv')
```

```python
# Lista de nombres válidos de Pokémon de acuerdo a los datos obtenidos
valid_pokemon = {
    'charizard', 'charmander', 'pikachu', 'scorbunny', 'rayquaza', 
    'piplup', 'blastoise', 'treecko', 'chandelure', 'dratini',
    'jigglypuff', 'empoleon', 'eevee', 'zorua', 'arcanine',
    'squirtle', 'totodile', 'mudkip', 'chikorita', 'lucario',
    'nidoking', 'blaziken', 'gengar', 'sprigatito', 'umbreon',
    "sirfetch'd", 'sableye', 'oshawott', 'sylveon', 'mimikyu',
    'snorlax', 'gardevoir', 'banette', 'haunter', 'togepi',
    'sandygast'
}
```

```python
# Diccionario de correcciones apreciables
corrections = {
    'picachu': 'pikachu',
    'evee': 'eevee',
    'tralalero tralala': None,
    'pokemon': None,
    'tu cosita rica': None,
    'vamo a calmarno': None,
    'no tengo mano': 'ninguno',
    'ni uno': 'ninguno',
    'todos': None,
    'x': None,
    'ninguno': 'ninguno',
    'no tengo': 'ninguno'
}
```

```python
def normalize_pokemon_name(name):
    # Convertir a minúsculas y quitar espacios extras
    name = str(name).lower().strip()
    
    # Correcciones específicas para variantes
    if name in corrections:
        return corrections[name]
    
    # Si el nombre ya está en la lista de válidos, retornarlo
    if name in valid_pokemon:
        return name.capitalize()
        
    return None
```

```python
# Aplicar la normalización
df['¿Cuál es tu Pokémon favorito?  '] = df['¿Cuál es tu Pokémon favorito?  '].apply(normalize_pokemon_name)
```
